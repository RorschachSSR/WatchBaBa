```r
# spot fig
subs <- data.frame(paste("sub",1:19,sep=""))
colnames(subs) = "subs"
allAvgDist <- bind_cols(subs,allAvgDist)
allexpAvgDist <- bind_cols(subs,allexpAvgDist)
coarse_seg_agree <- bind_cols(subs,coarse_seg_agree)
fine_seg_agree <- bind_cols(subs,fine_seg_agree)
creaPoint <- bind_cols(subs,creaPoint)
AvgDist <- bind_cols(subs,AvgDist)

spot_allAvgDist_Crea <- bind_cols(pivot_longer(allAvgDist,2:16,names_to="con",values_to="AvgDist"),
                                pivot_longer(creaPoint,2:16,names_to="con",values_to="creaPoint")[,3])
spot_allexpAvgDist_Crea <- bind_cols(pivot_longer(allexpAvgDist,2:16,names_to="con",values_to="allexpAvgDist"),
                                pivot_longer(creaPoint,2:16,names_to="con",values_to="creaPoint")[,3])
spot_coarse_seg_agree_Crea <- bind_cols(pivot_longer(coarse_seg_agree,2:16,names_to="con",values_to="coarse_seg_agree"),
                                pivot_longer(creaPoint,2:16,names_to="con",values_to="creaPoint")[,3])
spot_fine_seg_agree_Crea <- bind_cols(pivot_longer(fine_seg_agree,2:16,names_to="con",values_to="fine_seg_agree"),
                                pivot_longer(creaPoint,2:16,names_to="con",values_to="creaPoint")[,3])

spot_allAvgDist_Crea <- separate(spot_allAvgDist_Crea,con,sep="nvl",into=c("lvl","nvl"))
spot_allexpAvgDist_Crea <- separate(spot_allexpAvgDist_Crea,con,sep="nvl",into=c("lvl","nvl"))
spot_coarse_seg_agree_Crea <- separate(spot_coarse_seg_agree_Crea,con,sep="nvl",into=c("lvl","nvl"))
spot_fine_seg_agree_Crea <- separate(spot_fine_seg_agree_Crea,con,sep="nvl",into=c("lvl","nvl"))

spot_allAvgDist_Crea$AvgDist <- spot_allAvgDist_Crea$AvgDist/spot_allexpAvgDist_Crea$allexpAvgDist
spot_AvgDist_Crea <- spot_allAvgDist_Crea

plot_AvgDist_Crea <- ggplot(spot_AvgDist_Crea,aes(x=AvgDist,y=creaPoint, shape=nvl,color=nvl,group=lvl))+ 
                    geom_jitter(size=3,fill="white")
plot_coarse_seg_agree_Crea <- ggplot(spot_coarse_seg_agree_Crea,aes(x=coarse_seg_agree,y=creaPoint, shape=nvl,color=nvl,group=lvl))+ 
                    geom_jitter(size=3,fill="white")
plot_fine_seg_agree_Crea <- ggplot(spot_fine_seg_agree_Crea,aes(x=fine_seg_agree,y=creaPoint, shape=nvl,color=nvl,group=lvl))+ 
                    geom_jitter(size=3,fill="white")
```

```r
# corr fig for coarse split
Coatutudatapath <- ('C:/Users/ASUS/Desktop/WatchBaBa/WatchBaBaInMatlab/figure/data/GroupData')
Coatutu <- list()
for (i in 1:15){
  tutudataname <- file.path(Coatutudatapath,paste('Video_NO_',as.character(i),'_coarse.mat',sep=""))
  tutudata <- readMat(tutudataname)%>%
    data.frame()
  tutudata<-data.frame(t(tutudata))
  col_tutunames <- paste("sub",1:19,sep="")
  colnames(tutudata) <- col_tutunames
  tutudata <- mutate(tutudata,sum=rowSums(tutudata/19))
  sumtoeach <- data.frame()
  for(j in 1:19){
    sumtoeach_j <- Corr(bind_cols(tutudata[][20],tutudata[][j]),method = "spearman",plot='false')%>%
      as_md_tbl()%>%
      filter(.rownames=='sum'&.colnames!='sum')
    sumtoeach <- bind_rows(sumtoeach,sumtoeach_j)
  }
  prline <- sumtoeach%>%
    mutate(rd = cut(r, breaks = c(-Inf, 0.2, 0.4, Inf),
                    labels = c("< 0.3", "0.3 - 0.5", ">= 0.5")),
           pd = cut(p, breaks = c(-Inf, 0.01, 0.05, Inf),
                    labels = c("< 0.01", "0.01 - 0.05", ">= 0.05")))
  Coatutu[[i]] <-qcorrplot(correlate(tutudata[][1:19],method='spearman'), type = "upper", diag = FALSE) +
    geom_tile() +
    geom_mark(size=3.5,sig_level = c(0.05, 0.01, 0.001),
              mark = c("*", "**", "***"),sig_thres=0.05,sep="\n")+
    geom_couple(aes(colour = pd, size = rd), 
                data = prline, 
                curvature = nice_curvature(0.15),
                nudge_x = 0.2,
                label.fontface=2,
                label.size=4.5,
                drop=FALSE)+
    scale_fill_gradientn(colours = RColorBrewer::brewer.pal(9, "YlOrRd")) +
    scale_size_manual(values = c(0.5, 1, 2)) +
    scale_colour_manual(values = color_pal(3)) +
    guides(size = guide_legend(title = "cor",
                               override.aes = list(colour = "grey35"), 
                               order = 2),
           colour = guide_legend(title = "P_value", 
                                 override.aes = list(size = 3), 
                                 order = 1),
           fill = guide_colorbar(title = "Spearman's r", order = 3))+
    theme(axis.text=element_markdown(color="black",size=10),
          legend.background = element_blank(),
          legend.key = element_blank(),
          plot.title = element_text(hjust = 0.5, vjust = 0, size = 20))+
    labs(title = paste('corr for video level',as.character(ceiling(i/3)),'with novelty level', as.character(mod(i+1,3)+1),seq=""))
}
```

```r
# creaPoing GLM
for(i in 1:19){
  MAX <- max(creaPoint[i,])
  MIN <- min(creaPoint[i,])
  for (j in 1:15){
    creaPoint[i,j]=(creaPoint[i,j]-MIN)/(MAX-MIN)
  }
}

lmData <-bind_cols(pivot_longer(creaPoint,cols= 1:15,names_to = "video",values_to = "creaPoint"),
                   data.frame(allAvgDist=pivot_longer(allAvgDist,cols= 1:15,names_to = "video",values_to = "allAvgDist")$allAvgDist),
                   data.frame(allexpAvgDist=pivot_longer(allexpAvgDist,cols= 1:15,names_to = "video",values_to = "allexpAvgDist")$allexpAvgDist),
                   data.frame(coarse_seg_agree=pivot_longer(coarse_seg_agree,cols= 1:15,names_to = "video",values_to = "coarse_seg_agree")$coarse_seg_agree),
                   data.frame(fine_seg_agree=pivot_longer(fine_seg_agree,cols= 1:15,names_to = "video",values_to = "fine_seg_agree")$fine_seg_agree))%>%
  separate(.,video,sep="n",into=c("level","novel_level"))
lmData$level <- factor(lmData$level)
lmData$novel_level <- factor(lmData$novel_level,levels = c("vl1","vl2","vl3"),
                             labels = c("nvl1","nvl2","nvl3"))

Timedatapath <- ('C:/Users/ASUS/Desktop/WatchBaBa/WatchBaBaInMatlab/figure/data')
Timename <- file.path(datapath,'allTime.mat')
allTime <- readMat(Timename)%>%
  data.frame()
allTime <- allTime[,c(2,3,1,5,6,4,8,9,7,11,12,10,14,15,13)]%>%
  t()%>%
  data.frame()
colnames(allTime) <- "videoTime"
allTimes <- data.frame()
for (i in 1:19){
  allTimes <- bind_rows(allTimes,allTime)
}
lmData$videoTime <- allTimes$videoTime


lmData$hie_align <- lmData$allAvgDist/lmData$allexpAvgDist
L_model <- lm(creaPoint ~ hie_align+coarse_seg_agree+fine_seg_agree+level*novel_level*videoTime  , data = lmData)
```
